```python
import numpy as np
import matplotlib.pyplot as plt

limit = 100
step = 0.01

a, b, c, d, e = -12, -18, 5, 10, -30

x = np.arange(-limit, limit, step)

def func(x):
    return a * x**4 * np.sin(np.cos(x)) + b * x**3 + c * x**2 + d * x + e

plt.plot(x,func(x), 'b')
plt.grid()
plt.show()
```
